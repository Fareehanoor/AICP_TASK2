```python
# Initialize data for train journeys
train_schedule = {
    "09:00": {"available_seats": 480, "total_passengers": 0, "total_revenue": 0},
    "11:00": {"available_seats": 480, "total_passengers": 0, "total_revenue": 0},
    "13:00": {"available_seats": 480, "total_passengers": 0, "total_revenue": 0},
    "15:00": {"available_seats": 480, "total_passengers": 0, "total_revenue": 0},
    "10:00": {"available_seats": 480, "total_passengers": 0, "total_revenue": 0},
    "12:00": {"available_seats": 480, "total_passengers": 0, "total_revenue": 0},
    "14:00": {"available_seats": 480, "total_passengers": 0, "total_revenue": 0},
    "16:00": {"available_seats": 640, "total_passengers": 0, "total_revenue": 0},
}

def purchase_tickets():
    departure_time = input("Enter departure time (e.g., 09:00): ")
    return_time = input("Enter return time (e.g., 10:00): ")
    num_tickets = int(input("Enter the number of tickets: "))

    if departure_time in train_schedule and return_time in train_schedule:
        journey = train_schedule[departure_time]
        if journey["available_seats"] >= num_tickets:
            total_price = num_tickets * 25
            discount = (num_tickets // 10) * 25
            total_price -= discount

            journey["available_seats"] -= num_tickets
            journey["total_passengers"] += num_tickets
            journey["total_revenue"] += total_price

            print(f"Tickets purchased successfully. Total price: ${total_price}")
        else:
            print("Not enough available seats for the requested tickets.")
    else:
        print("Invalid train time. Please choose a valid time.")

def end_of_day_summary():
    total_passengers = sum(journey["total_passengers"] for journey in train_schedule.values())
    total_revenue = sum(journey["total_revenue"] for journey in train_schedule.values())

    max_passengers_journey = max(train_schedule, key=lambda k: train_schedule[k]["total_passengers"])

    print("End of Day Summary:")
    for time, info in train_schedule.items():
        print(f"{time} - Total Passengers: {info['total_passengers']}, Total Revenue: ${info['total_revenue']}")

    print(f"Total Passengers for the Day: {total_passengers}")
    print(f"Total Revenue for the Day: ${total_revenue}")
    print(f"Train with the most passengers: {max_passengers_journey}")

while True:
    print("\nElectric Mountain Railway Menu:")
    print("1. Purchase Tickets")
    print("2. End of the Day")
    print("3. Exit")

    choice = input("Enter your choice (1/2/3): ")

    if choice == "1":
        purchase_tickets()
    elif choice == "2":
        end_of_day_summary()
    elif choice == "3":
        print("Goodbye!")
        break
    else:
        print("Invalid choice. Please choose a valid option (1/2/3).")
```
